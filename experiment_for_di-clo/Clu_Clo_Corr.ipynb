```python
import networkx as nx
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Soc_Alpha = nx.read_edgelist("datasets/BTC-Alpha", delimiter = ',', create_using=nx.DiGraph, data=(('weight', int),('sec', int),))
Soc_OTC = nx.read_edgelist("datasets/BTC-OTC", delimiter = ',', create_using=nx.DiGraph, data=(('weight', int),('sec', float),))
Soc_Advogato = nx.read_edgelist("datasets/Soc_Advogato.csv", delimiter = ',', create_using=nx.DiGraph, data=(('weight', float),))
Soc_Epinions = nx.read_edgelist("datasets/Soc_Epinions.csv", delimiter = ',', create_using=nx.DiGraph)

FW_Mangwet = nx.read_edgelist("datasets/FW-Mangwet", create_using=nx.DiGraph, data = (('weight', float),))
FW_Baywet = nx.read_edgelist("datasets/FW-Baywet", create_using=nx.DiGraph, data = (('weight', float),))
FW_Baydry = nx.read_edgelist("datasets/FW-Baydry", create_using=nx.DiGraph, data = (('weight', float),))
FW_LittleRock = nx.read_edgelist("datasets/FW-LittleRock", create_using=nx.DiGraph)

Cit_Cora = nx.read_edgelist("datasets/Cit-Cora", create_using=nx.DiGraph)
Cit_HepTh = nx.read_edgelist("datasets/Cit-HepTh", create_using=nx.DiGraph)
Cit_HepPh = nx.read_edgelist("datasets/Cit-HepPh", create_using=nx.DiGraph)
Cit_Citeseer = nx.read_edgelist("datasets/Cit-Citeseer", delimiter = ',', create_using=nx.DiGraph)

SW_Weka = nx.read_edgelist("datasets/SW_Weka", create_using=nx.DiGraph)
SW_Lucene = nx.read_edgelist("datasets/SW_Lucene", create_using=nx.DiGraph, data=(('weight', str),))
SW_Jung = nx.read_edgelist("datasets/SW_Jung.csv", delimiter = ',', create_using=nx.DiGraph)
SW_Jdk = nx.read_edgelist("datasets/SW_Jdk.csv", delimiter = ',', create_using=nx.DiGraph)

Web_Stanford = nx.read_edgelist("datasets/web-Stanford.txt", create_using=nx.DiGraph)
Web_NotreDame = nx.read_edgelist("datasets/web-NotreDame.txt", create_using=nx.DiGraph)
Web_BerkStan = nx.read_edgelist("datasets/web-BerkStan.txt", create_using=nx.DiGraph)
Web_Google = nx.read_edgelist("datasets/web-Google.txt", create_using=nx.DiGraph)

WordAdj_Jp = nx.read_edgelist("datasets/WordAdj_Jp.csv", delimiter = ',', create_using=nx.DiGraph)
WordAdj_Drw = nx.read_edgelist("datasets/WordAdj_Drw.csv", delimiter = ',', create_using=nx.DiGraph)
WordAdj_Fr = nx.read_edgelist("datasets/WordAdj_Fr.csv", delimiter = ',', create_using=nx.DiGraph)
WordAdj_Spa = nx.read_edgelist("datasets/WordAdj_Spa.csv", delimiter = ',', create_using=nx.DiGraph)
```

```python
df_btcAlpha = nx.get_cc_ce_df(Soc_Alpha)
df_btcOTC = nx.get_cc_ce_df(Soc_OTC)
df_advogato = nx.get_cc_ce_df(Soc_Advogato)
df_epinions = pd.read_csv("saved_clu_clo/epinions.csv", index_col=0)

df_mang = nx.get_cc_ce_df(FW_Mangwet)
df_baywet = nx.get_cc_ce_df(FW_Baywet)
df_baydry = nx.get_cc_ce_df(FW_Baydry)
df_littlerock = nx.get_cc_ce_df(FW_LittleRock)

df_cora = pd.read_csv("saved_clu_clo/cora.csv", index_col=0)
df_hepTh = pd.read_csv("saved_clu_clo/hepTh.csv", index_col=0)
df_hepPh = pd.read_csv("saved_clu_clo/hepPh.csv", index_col=0)
df_citeseer = pd.read_csv("saved_clu_clo/citeseer.csv", index_col=0)

df_weka = nx.get_cc_ce_df(SW_Weka)
df_lucene = nx.get_cc_ce_df(SW_Lucene)
df_jung = nx.get_cc_ce_df(SW_Jung)
df_jdk = nx.get_cc_ce_df(SW_Jdk)

df_stanford = pd.read_csv("saved_clu_clo/stanford.csv", index_col=0)
df_notredame = pd.read_csv("saved_clu_clo/notredame.csv", index_col=0)
df_berkstan = pd.read_csv("saved_clu_clo/berkstan.csv", index_col=0)
df_google = pd.read_csv("saved_clu_clo/google.csv", index_col=0)

df_japanese = nx.get_cc_ce_df(WordAdj_Jp)
df_english = nx.get_cc_ce_df(WordAdj_Drw)
df_french = nx.get_cc_ce_df(WordAdj_Fr)
df_spanish = nx.get_cc_ce_df(WordAdj_Spa)
```

```python
def print_corr(x_y, method):
    print("---------Method: {}-----------\n".format(method))
    print("Tr-BTCAlpha: {:.3f}\n".format(df_btcAlpha[x_y].corr(method=method).iloc[0][1]))
    print("Tr-BTC-OTC: {:.3f}\n".format(df_btcOTC[x_y].corr(method=method).iloc[0][1]))
    print("Tr-Advogato: {:.3f}\n".format(df_advogato[x_y].corr(method=method).iloc[0][1]))
    print("Tr-Epinions: {:.3f}\n".format(df_epinions[x_y].corr(method=method).iloc[0][1]))

    print("FW-Mangrove: {:.3f}\n".format(df_mang[x_y].corr(method=method).iloc[0][1]))
    print("FW-BayWet: {:.3f}\n".format(df_baywet[x_y].corr(method=method).iloc[0][1]))
    print("FW-BayDry: {:.3f}\n".format(df_baydry[x_y].corr(method=method).iloc[0][1]))
    print("FW-LittleRock: {:.3f}\n".format(df_littlerock[x_y].corr(method=method).iloc[0][1]))

    print("Cit-Cora: {:.3f}\n".format(df_cora[x_y].corr(method=method).iloc[0][1]))
    print("Cit-HepTh: {:.3f}\n".format(df_hepTh[x_y].corr(method=method).iloc[0][1]))
    print("Cit-HepPh: {:.3f}\n".format(df_hepPh[x_y].corr(method=method).iloc[0][1]))
    print("Cit-Citeseer: {:.3f}\n".format(df_citeseer[x_y].corr(method=method).iloc[0][1]))

    print("SW-Weka: {:.3f}\n".format(df_weka[x_y].corr(method=method).iloc[0][1]))
    print("SW-Lucene: {:.3f}\n".format(df_lucene[x_y].corr(method=method).iloc[0][1]))
    print("SW-JUNG: {:.3f}\n".format(df_jung[x_y].corr(method=method).iloc[0][1]))
    print("SW-JDK: {:.3f}\n".format(df_jdk[x_y].corr(method=method).iloc[0][1]))

    print("Web-Stanford: {:.3f}\n".format(df_stanford[x_y].corr(method=method).iloc[0][1]))
    print("Web-NotreDame: {:.3f}\n".format(df_notredame[x_y].corr(method=method).iloc[0][1]))
    print("Web-BerkStan: {:.3f}\n".format(df_berkstan[x_y].corr(method=method).iloc[0][1]))
    print("Web-Google: {:.3f}\n".format(df_google[x_y].corr(method=method).iloc[0][1]))

    print("WA-Japanese: {:.3f}\n".format(df_japanese[x_y].corr(method=method).iloc[0][1]))
    print("WA-English: {:.3f}\n".format(df_english[x_y].corr(method=method).iloc[0][1]))
    print("WA-French: {:.3f}\n".format(df_french[x_y].corr(method=method).iloc[0][1]))
    print("WA-Spanish: {:.3f}\n".format(df_spanish[x_y].corr(method=method).iloc[0][1]))
```

```python
print_corr(['cc','ce'], 'pearson')
```

```python
print_corr(['cc','ce'], 'spearman')
```

```python
print_corr(['cc','ce'],"kendall")
```
